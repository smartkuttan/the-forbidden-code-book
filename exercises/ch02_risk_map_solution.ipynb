```python
import matplotlib.pyplot as plt

risks = {
    "Bias in dataset": (0.8, 0.9),
    "Privacy violation": (0.7, 1.0),
    "System downtime": (0.6, 0.4),
    "Deceptive design": (0.9, 0.7)
}

plt.figure(figsize=(6,5))
for name, (l, i) in risks.items():
    plt.scatter(l, i, s=80)
    plt.text(l+0.01, i+0.01, name, fontsize=8)
plt.xlim(0,1); plt.ylim(0,1)
plt.xlabel("Likelihood (0–1)"); plt.ylabel("Impact (0–1)")
plt.title("AI Ethics Risk Map")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
